# Task 1: User Login Check

```python
username = "admin"
password = "1234"

input_user = input("Enter username: ")
input_pass = input("Enter password: ")

if input_user == username and input_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# Task 2: Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
user_marks = int(input("Enter your marks: "))
if user_marks >= 50:
    print("You have Passed")
else:
    print("You have Failed")
```

# Task 3: Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned_names.append(name.strip().lower())
print("Cleaned Names:", cleaned_names)
```

# Task 4: Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length = len(message)
    print(f"Message: '{message}' → Length: {length}")

    if length > 10:
        print("⚠️ Message is longer than 10 characters")
```
